# Dimensionality Reduction

Real-world datasets often have too many dimensions to fit into memory, be processed quickly, or allow statistical inference; therefore, we must consider only some.

Consider the iris dataset for example.

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

## Manual feature selection

The simplest way to consider fewer dataset dimensions is to choose some ourselves.

```python
import matplotlib.pyplot as plt
```

For example, we can choose two iris features: sepal length and width.

```python
X = iris.data[:, :2]
y = iris.target
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

Plotting the data along them reveals an elegantly expressive graph.

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')

plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks()
plt.yticks()

plt.show()
```

## Principal Component Analysis

Or, we can use let an algorithm determine which feature combinations would best constitute a user-chosen number of axes.

```python
from sklearn.decomposition import PCA
X_reduced = PCA(n_components=3).fit_transform(iris.data)
```

Plotting this reduced dataset in its three dimensions reveals even more information.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8, 6))

ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, 
           cmap=plt.cm.Set1, edgecolor='k', s=40)

ax.set_title("Principal Component Analysis")
ax.set_xlabel("1st")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
